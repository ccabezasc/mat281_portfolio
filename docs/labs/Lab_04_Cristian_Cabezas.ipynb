Lo primero es cargar el conjunto de datos y ver las primeras filas que lo componen:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(10,8)})
```

```python
# cargar datos
df = pd.read_csv("https://raw.githubusercontent.com/fralfaro/MAT281_2022/main/labs/lab_04/data/company_sales_data.csv")
df.head()
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

**Observación.-** Puedes ocupar las librerías de Matplolib o Seaborn.

1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
# Desarrollo:

Month = df['month_number']
Total_Profit = df['total_profit']

fig = plt.figure(figsize=(10, 5))

plt.plot(
    Month, # eje x
    Total_Profit, # eje y
    marker = 'o', # "dispersion"
    color="black", # color
    linewidth=1 # tamano de la curva
)
plt.legend() # agregar etiquetado
plt.title("grafico simple") # agregar titulo
plt.xlabel('Month') # nombre eje x
plt.ylabel('Total_Profit') # nombre eje y
#plt.grid() # agregar grillado
plt.show() # mostrar grafico

```

2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
# Desarrollo

products = df.columns[1:7]

fig, ax = plt.subplots(figsize=(10,6))


for product in products:
  sales = df[product]
  ax.plot(df['month_number'], sales, marker = 'o', label = product)

ax.set_xlabel('Mes')
ax.set_ylabel('Ventas')
ax.set_title('Ventas mensuales')
ax.legend()
plt.grid(1)
plt.show()



```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
# Desarrollo
facecream_sales = df['facecream']
facewash_sales = df['facewash']
months = df['month_number']

bar_width = 0.35
x= range(len(months))

plt.bar(x, facecream_sales, width = bar_width, label='Facecream', align='center')
plt.bar([i + bar_width for i in x], facewash_sales, width = bar_width, label='Facewash', align='center')

# Etiquetas y título
plt.xlabel('Meses')
plt.ylabel('Ventas')
plt.title('Comparación de Ventas de Productos')
plt.xticks([i + bar_width/2 for i in x], months)
plt.legend()

# Mostrar el gráfico
plt.show()
```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
# Desarrollo

product_data = df.iloc[:,1:7]

fig, ax = plt.subplots(figsize=(10,6))

ax.boxplot(product_data.values)

ax.set_xticklabels=(df.columns[1:7])
ax.set_xlabel('Productos')
ax.set_ylabel('Ventas')
ax.set_title('Distribucion de ventas')

plt.grid(1)
plt.show()
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
# Desarrollo
last_year_data = df[df['month_number']>= 12]
product_sales = last_year_data.iloc[:, 1:7].sum()

fig, ax = plt.subplots(figsize=(10,6))
ax.pie(product_sales, labels = product_sales.index, autopct = '%1.1f%%', startangle = 90)
ax.set_title('Ventas totales por porducto (Año pasado)')
plt.show()
```
